```python
import os

# Replace 'data_dir' with the path to your data directory
data_dir = '/Users/wolf7031/Documents/Quran/RGB ArSL dataset'

# Get the class labels from the subdirectory names
class_names = [d.name for d in os.scandir(data_dir) if d.is_dir()]
class_names.sort()  # Sort the class names for consistent ordering

print("Class Labels:")
for idx, class_name in enumerate(class_names):
    print(f"{idx}: {class_name}")
```
